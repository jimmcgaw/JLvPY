```julia
using CSV
using Flux
using DataFrames
using Plots
```

```julia
training_data = CSV.read("./src/mnistdigits/mnist_train.csv", DataFrame)
test_data = CSV.read("./src/mnistdigits/mnist_test.csv", DataFrame)
```

```julia
train_y = training_data[!, 1]
train_x = training_data[!, 2:785]
test_y = test_data[!, 1]
test_x = test_data[!, 2:785]
size(test_x)
```

```julia
# handy example: https://fluxml.ai/Flux.jl/stable/guide/models/quickstart/
model = Chain(
    Dense(784, 1),
    Dense(128, 1, NNlib.relu),
    Dense(10, 1, NNlib.softmax),
)
optimizer = Flux.Adam(0.001)
```

```julia
loss = Flux.Losses.crossentropy(test_x, test_y)
```

```julia
# Training loop, using the whole data set 10 times:
epochs = 10
for epoch in 1:epochs
    Flux.train!(Flux.Losses.crossentropy, Flux.params(model), [train_x, train_y], optimizer)
    println("Epoch $epoch: Loss = $(loss(X_train, y_train))")
end
```
